```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

mu = 0  
sigma = 1 
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

z_critical_alpha_005 = -1.645  # Updated z-score for alpha level of 0.05

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='navy', linewidth=2)

x_fill_alpha_005 = np.linspace(z_critical_alpha_005, -(mu + 3*sigma), 1000)
y_fill_alpha_005 = stats.norm.pdf(x_fill_alpha_005, mu, sigma)
plt.fill_between(x_fill_alpha_005, y_fill_alpha_005, 0, alpha=0.3, color='orange', label='Rejection Region ($\\alpha=0.05$)')

plt.axvline(z_critical_alpha_005, color='red', linestyle='--', label='Critical Value ($z=-1.645$)')
plt.title('Standard Normal Distribution with One-Tailed Hypothesis Test ($\\alpha=0.05$)', fontsize=14)
plt.xlabel('Standard Deviations from Mean', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()

```

```python
z_critical_two_tailed = 1.96  

x_fill_left = np.linspace(mu - 3*sigma, -z_critical_two_tailed, 1000)
y_fill_left = stats.norm.pdf(x_fill_left, mu, sigma)
x_fill_right = np.linspace(z_critical_two_tailed, mu + 3*sigma, 1000)
y_fill_right = stats.norm.pdf(x_fill_right, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='navy', linewidth=2)

plt.fill_between(x_fill_left, y_fill_left, 0, alpha=0.3, color='orange', label='Rejection Region (Left Tail)')
plt.fill_between(x_fill_right, y_fill_right, 0, alpha=0.3, color='orange', label='Rejection Region (Right Tail)')

plt.axvline(-z_critical_two_tailed, color='red', linestyle='--', label='Critical Value ($z=-1.96$)')
plt.axvline(z_critical_two_tailed, color='red', linestyle='--', label='Critical Value ($z=1.96$)')

plt.title('Standard Normal Distribution with Two-Tailed Hypothesis Test ($\\alpha=0.05$)', fontsize=14)
plt.xlabel('Standard Deviations from Mean', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```
